# Extract frames from video

```python
num_frame = 15
violence_path = "/kaggle/input/real-life-violence-situations-dataset/"
non_violence_path = "/kaggle/input/real-life-violence-situations-dataset/"
```

```python
os.makedirs('./data/Violence',exist_ok=True)
for path in tqdm(glob.glob(violence_path+'/*')):
    fname = os.path.basename(path).split('.')[0]
    vidcap = cv2.VideoCapture(path)
    success,image = vidcap.read()
    count = 0
    while success:
        if count % num_frame == 0: #Take 1 frame every num_frame
            cv2.imwrite("./data/Violence/{}-{}.jpg".format(fname,str(count).zfill(4)),image)     # save frame as JPEG file      
        success,image = vidcap.read()
        count += 1
```

```python
os.makedirs('./data/NonViolence',exist_ok=True)
for path in tqdm(glob.glob(non_violence_path+'/*')):
    fname = os.path.basename(path).split('.')[0]
    vidcap = cv2.VideoCapture(path)
    success,image = vidcap.read()
    count = 0
    while success:
        if count % num_frame == 0:
            cv2.imwrite("./data/NonViolence/{}-{}.jpg".format(fname,str(count).zfill(4)),image)     # save frame as JPEG file      
        success,image = vidcap.read()
        count += 1
```

```python
# Step 1: Create the folder "train" if it doesn't exist
output_folder = "train"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Step 2: Create folders for class 0 and class 1
class0_folder = os.path.join(output_folder, "class0")
class1_folder = os.path.join(output_folder, "class1")
if not os.path.exists(class0_folder):
    os.makedirs(class0_folder)
if not os.path.exists(class1_folder):
    os.makedirs(class1_folder)

# Step 3: Iterate through the NumPy arrays X and Y and save each image in the corresponding class folder
for i, (image_data, label) in enumerate(zip(trainX, trainY)):
    image = Image.fromarray(np.uint8(image_data))
    if label[0] == 1.0:  # Class 0
        image_path = os.path.join(class0_folder, f"image_{i}.png")
    else:  # Class 1
        image_path = os.path.join(class1_folder, f"image_{i}.png")
    image.save(image_path)

print("Images saved successfully in the 'train' folder.")
```
